```python
# load a dataframe from csv file to prep for analysis

import pandas as pd

df = pd.read_csv('data/annotations 12-29.csv')
df
```

```python
# make a df for carolina mentions and another for all other brands
# carolina: brand column contains 'carolina' ignore case
# flinn: brand column contains 'flinn' ignore case
# amazon: brand column contains 'amazon' ignore case

carolina = df[df['brand'].str.contains('carolina', case=False)]
flinn = df[df['brand'].str.contains('flinn', case=False)]
amazon = df[df['brand'].str.contains('amazon', case=False)]
non_carolina = df[~df['brand'].str.contains('carolina', case=False)]
# filter empty cells and '[]' value
non_carolina = non_carolina[non_carolina['brand'].notnull()]
non_carolina = non_carolina[non_carolina['brand'] != '[]']

# write each file to csv
carolina.to_csv('data/carolina.csv')
flinn.to_csv('data/flinn.csv')
amazon.to_csv('data/amazon.csv')
non_carolina.to_csv('data/non_carolina.csv')
```

```python
# make a frequency table for each brand
# only use the text_segment column for the frequency table

# carolina word frequency
carolina_freq = pd.Series(' '.join(carolina['text_segment']).split()).value_counts()
carolina_freq
```
